```python
tripadvisor = pd.read_csv("data/lasvegas_tripadvisor.csv")
tripadvisor.columns
```

## Bar Plots

```python
# Grafico Común
tripadvisor['Traveler type'].value_counts().plot.bar()
```

```python
sns.set(font_scale=1.3)
sns.set_style("darkgrid")
sns.catplot(data=tripadvisor, x="Traveler type" , y="Score", kind="bar", hue="Traveler type")
```

```python
tripadvisor['Traveler type'] = tripadvisor['Traveler type'].astype('category')
tripadvisor['Traveler type'].cat.categories
```

```python
sns.catplot(data=tripadvisor, x="Traveler type" , y="Score", kind="bar", hue="Tennis court")
```

```python
tripadvisor['Period of stay'].value_counts()
```

```python
sns.catplot(data=tripadvisor, y="Helpful votes", x='Period of stay', hue='Period of stay', kind='bar')
```

```python
continent_categories = list(tripadvisor["User continent"].value_counts().index)
tripadvisor['User continent'] = tripadvisor['User continent'].astype('category')
tripadvisor["User continent"] = tripadvisor["User continent"].cat.reorder_categories(continent_categories)
sns.set(font_scale=.8)
sns.catplot(data=tripadvisor, x="User continent", y="Score", kind="bar", hue='User continent')
```

## Point and Count plots

```python
sns.catplot(data=tripadvisor, x="Pool", y='Score', kind="point")
```

```python
sns.catplot(data=tripadvisor, x="Spa", y='Score', kind="point", hue='Tennis court', dodge=True)
```

```python
sns.catplot(data=tripadvisor , x='Score' , y='Review weekday', kind='point', join=False)
```

```python
sns.catplot(data=tripadvisor , x='Tennis court', kind='count' , hue='Spa')
```

## Opciones adicionales

```python
ax = sns.catplot(x="Traveler type", data=tripadvisor, kind="count", col='User continent',col_wrap=3,palette=sns.color_palette('Set1'))
ax.fig.suptitle('Traveler type per user continent')
ax.set_axis_labels("Traveler type", "Count")
plt.subplots_adjust(top=0.9)
plt.show()
```
