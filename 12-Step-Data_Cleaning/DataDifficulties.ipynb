```python
import pandas as pd

ride_sharing = pd.read_csv("data/ride_sharing_new.csv", index_col=0)

ride_sharing.info()
```

```python

ride_sharing['user_type'].describe()
```

```python
ride_sharing['user_type_cat'] = ride_sharing['user_type'].astype('category')
assert ride_sharing['user_type_cat'].dtype == 'category'
ride_sharing['user_type_cat'].describe()
```

```python
ride_sharing['duration_trim'] = ride_sharing['duration'].str.strip('minutes')
ride_sharing['duration_time'] = ride_sharing['duration_trim'].astype('int')
assert ride_sharing['duration_time'].dtype == 'int'


print(ride_sharing['duration_time'].mean())
```

Duplicados => 
* data.duplicated()
    - 2 argumentos claves: subset (columnas), keep (first, last, False)
* data.drop_duplicates()
    - 3 argumentos claves: subset (columnas), keep (first, last, False), inplace (True, False)
