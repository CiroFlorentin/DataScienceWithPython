# Importing flat files 

## Numpy

```python
import numpy as np
import matplotlib.pyplot as plt

filename = 'data/MNIST_train.txt'

digits = np.loadtxt(filename, delimiter=',') #skiprows=1 dtype=int usecols=[0,2]

digits
```

```python
im = digits[21, 1:]
im_sq = np.reshape(im, (28, 28))

plt.imshow(im_sq, cmap='Greys', interpolation='nearest')
plt.show()
```
